```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load the dataset
file_path = "/content/drive/MyDrive/UpdatedDataset/Dataset(2014-2024).csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# ----- Encode 'Crop' -----
crop_encoder = OneHotEncoder(sparse_output=False, drop=None)
crop_encoded = crop_encoder.fit_transform(df[['Crop']])
crop_encoded_df = pd.DataFrame(crop_encoded, columns=crop_encoder.get_feature_names_out(['Crop']))

# ----- Encode 'District' -----
district_encoder = OneHotEncoder(sparse_output=False, drop=None)
district_encoded = district_encoder.fit_transform(df[['District']])
district_encoded_df = pd.DataFrame(district_encoded, columns=district_encoder.get_feature_names_out(['District']))

# Combine all parts: drop original columns and concatenate new encoded ones
df_encoded = pd.concat([df.drop(columns=['Crop', 'District']), crop_encoded_df, district_encoded_df], axis=1)

# Show result
print(df_encoded.head())
```

```python
# Save the encoded DataFrame to a CSV file
output_path = "all_data_encoded.csv"
df_encoded.to_csv(output_path, index=False)

print(f"✅ Encoded dataset saved to: {output_path}")
```
